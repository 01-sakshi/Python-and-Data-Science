### Red Wine Data through Numpy

```python
import numpy as np
wine = np.genfromtxt("Documents/winequality-red.csv", delimiter=";", skip_header=1)
print(type(wine))
print("\nWines data:\n",wine)
```

```python
#Questions:
print("Size of array:",wines.size)
row=len(wines)
col=len(wines[0])
print("No of rows:",row)
print("No of cols:",col)
print("Shape of array:",wines.shape)

print(wines[:,11])      #Quality Column
print(wines[:,11]+2)    #Adding 2 to the quality column
```

```python
#See the rows where the quality is b/w 5 and 6.
qual_col=wines[:,11]
res_col=[]
#print(type(res_col))
for val in qual_col:
    if 5<=val<=6:
        res_col.append(val)
        #print(res_col)
print(np.array(res_col))
```

```python
#Add row-wise the values and put it in an array.
import numpy as np
from numpy import *
list_sum=[]
for i in range(0,row):
    res=around(wines[i,:].sum(),2)   #around() is present in numpy
    list_sum.append(res)
print(np.array(list_sum))
```

```python
#Add rows 1 to 4.
import numpy as np
from numpy import *
list_sum=[]
sum=0
for i in range(0,4):
    res=around(wines[i,:].sum(),2)   #around() is present in numpy
    sum+=res
    list_sum.append(sum)
print(np.array(list_sum))
```

```python
#To see wines which have a quality rating equalto 10.
qual_col=wine[:,11]
res_col=[]
for val in range(len(qual_col)):
    if qual_col[val]==7:
        #np.set_printoptions(formatter={'float_kind':'{:.1f}'.format})
        print("wine:",wine[val,:])
    else:
        pass
print("\nSee only rows 0 to 2:")
print("wine:\n",wine[:3,:])
```

```python
qual=wines[:,11]
print("Sum:",qual.sum())   #sum of the quality column
print("Average without using axis:",round(average(qual),2))  #Average without using axis
```
