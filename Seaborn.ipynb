## Seaborn

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns

#Relational Plot
tips=sns.load_dataset("tips")
sns.relplot(x="total_bill",y="tip",data=tips)
```

Types of Relational Plot:
    1. Scatter Plot (scatterplot() by default)
    2. Line Plot    (lineplot())

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker")
```

```python
sns.relplot(x="total_bill",y="tip",data=tips,hue="smoker",style="smoker")
#to add marker style to smoker
```

Types of Categorical Plots:
    1. Scatter Plots       (stripplot() by default , swarmplot())
    2. Distribution Plots  (boxplot() , violinplot() , boxenplot())
    3. Estimate Plots       (pointplot() , barplot() , countplot())

```python
#Categorical Plot
  #Scatter Plot
res=sns.catplot(x="day",y="total_bill",data=tips)
type(res)
print(res)
```

```python
tips['day'].unique()
```

```python
#by default 'jitter' is set to True.
sns.catplot(x="day",y="total_bill",jitter=False,data=tips)
# jitter=False reduces the scattering.
```

```python
sns.catplot(x="day",y="total_bill",hue="time",kind="swarm",data=tips)
```

```python
sns.catplot(x="smoker",y="tip",data=tips)
#sns.catplot(x="smoker",y="tip",order=["No","Yes"],data=tips)
```

Distribution Plots:
    1. Box Plot    ( boxplot() )
    2. Violin Plot ( violinplot() )
    3. boxenplot()

```python
#Distribution Plots
  #Box Plot
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.boxplot(x="time",y="size",data=tips)
#box plot is applied on numerical values
```
