## Matplotlib

```python
from matplotlib import pyplot as plt
%matplotlib inline

#By default plot() func creates a Line Chart
plt.plot([1,2,3],[4,5,6])
plt.locator_params(integer=True)
#plotted points- (1,4) (2,5) (3,6)
plt.show()
```

```python
plt.plot([1,2,3],[3,5,10])
#plt.xticks([1,2,3])
plt.locator_params(integer=True)
plt.show()
```

```python
from matplotlib import style
x=[5,8,10]
y=[12,16,6]
plt.plot(x,y)
plt.title('My Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```

```python
#Line Chart
x1=[1,3,5]
y1=[2,4,6]

x2=[7,6,9]
y2=[4,7,10]

plt.plot(x1,y1,'g',label='Line 1',linewidth=3)
plt.plot(x2,y2,'r',label='Line 2',linewidth=3)
plt.legend(loc='best')
```

```python
#Marker
x1=[1,3,5]
y1=[2,4,6]

x2=[7,6,9]
y2=[4,7,10]

plt.plot(x1,y1,'--gD',label='Line 1',linewidth=3,alpha=0.5)    #'alpha' is for transparency
plt.plot(x2,y2,'--r+',label='Line 2',linewidth=3)
plt.legend(loc='best')
```

```python
#Bar Chart
import numpy as np

company=['Facebook','Google','Flipkart','Zomato','Ola']
revenue=[125,65,72,88,101]
xpos=np.arange(len(company))   #To show numerical value in place of String value in x-axis
plt.xticks(xpos,company)
plt.bar(xpos,revenue)
plt.show()
```

```python
#Bar Chart
profit=[120,43,56,111,57]
revenue=[125,65,72,88,101]
plt.title('Two Bars')
plt.bar(xpos-0.2,profit,width=0.4,label='Profit')
plt.bar(xpos+0.2,revenue,width=0.4,label='Revenue')
#'width' is the width of bar
plt.legend()
```

```python
#Histogram
#Histograms are bar charts where y-axis contains frequency of variables.
blood_sugar=[120,210,78,95]
#bins=[80,110,140,170,200]
plt.hist(blood_sugar)
```

```python
#Pie-Chart
exp_values=[30,45,90,120]
exp_labels=['Phoebe','Rachel','Chandler','Joey']
explode=[0.2,0,0,0]
plt.axis('equal')
plt.pie(exp_values,explode,labels=exp_labels,radius=1.5,shadow=True)
```

```python
#1.  Display foot preference of FIFA players
#df.columns

from matplotlib import pyplot as plt
%matplotlib inline

L,R=df['Preferred Foot'].value_counts()
#print(type(L))

lisp=[L,R]
#plt.hist(lisp,rwidth=0.9)
show_labels=['Left Foot','Right Foot']
plt.pie(lisp,labels=show_labels,autopct='%.2f%%')
```

```python
#2. FInd out distribution of players skills in Fifa 2018

#df['Skill Moves'].unique()
a,b,c,d,e=df['Skill Moves'].value_counts()
#print(a,b,c,d,e)
skill=[a,b,c,d,e]
plt.hist(skill,rwidth=0.9,bins=[50,950,2500,6600,9000])
```

```python
#3. Display gas prices over time in USD
from matplotlib import pyplot as plt
%matplotlib inline
import pandas as pd

df1=pd.read_csv("gas-prices.csv")
plt.bar(df1['Year']+0.2,df1['USA'],width=0.4,color="green", label="USA")
plt.bar(df1['Year']+0.4,df1['Germany'],width=0.4,color="red", label="Canada")
plt.bar(df1['Year']-0.2,df1['UK'],width=0.4,color="yellow", label="UK")

plt.xticks(df1.Year[::4])
plt.xlabel("Year")
plt.ylabel("Price")
plt.legend()
plt.show()
```
